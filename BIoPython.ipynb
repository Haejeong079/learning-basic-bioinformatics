```python
from Bio.Seq import Seq

# DNA 서열 생성
dna_seq = Seq("ATGCGTACGTACG")
print(dna_seq)  # 출력: ATGCGTACGTACG
print(len(dna_seq))  # 서열 길이: 13
                        
```

```python
def calculate_gc_content (dna_seq):
    g_count = dna_seq.count('G')
    c_count = dna_seq.count('C')
    total_count = len(dna_seq)
    gc_count = (g_count + c_count) / total_count * 100
    return gc_count

print(f"GC Content: {calculate_gc_content(dna_seq):.2f}%")
```

```python
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq

dna_seq = Seq("ATGCGTACGTAGCTAGCTAG")

# SeqRecord 생성
record = SeqRecord(
    seq=dna_seq,
    id='gene123',
    name='example_gene',
    description='An example gene sequence'
)


print(record.id)
print(record.name)
print(record.description)
print(record.seq)
```

```python
# 주석 정보 추가 하기
record.annotations ["organism"] = "Homo sapiens"
record.annotations ["molecule_type"] = "DNA"
```

```python
# 특징 (feature) 추가 하기
from Bio.SeqFeature import SeqFeature, FeatureLocation

feature = SeqFeature(
    FeatureLocation(start=0, end=3),
    type="CDS",
    qualifiers={"gene": "example_gene"}
)
record.features.append(feature)

print(record.annotations["organism"])
print(record.features[0].type)
```

```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

records=[]
for i in range(3):
    seq= Seq(f"ATGC{'A'*i}GTAC")
    record = SeqRecord(seq, id=f"example_{i}", description=f"Sequence{i+1}")
    records.append(record)


SeqIO.write(records, "output.fasta", "fasta")
print("FASTA 파일이 생성되었습니다.")
```

```python
# 필요한 라이브러리 불러오기
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction   # Biopython 1.79 이상
import matplotlib.pyplot as plt

# FASTA 파일 읽기 (multi-FASTA)
records = SeqIO.parse("output.fasta", "fasta")

# 각 서열의 ID와 GC% 저장
ids = []
gc_values = []

for record in records:
    ids.append(record.id)
    gc_values.append(gc_fraction(record.seq) * 100)  # 0~1 → 0~100%

# 막대그래프 시각화
plt.figure(figsize=(7, 4))
plt.bar(ids, gc_values, color='seagreen', edgecolor='black')

plt.title("GC Content Comparison", fontsize=14)
plt.xlabel("Sequence ID", fontsize=12)
plt.ylabel("GC Content (%)", fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
# FASTA 파일에서 여러 서열 읽기
records = list(SeqIO.parse("output.fasta", "fasta"))

# 각 서열의 길이 계산
lengths = [len(record.seq) for record in records]

# histogram 시각화
plt.rcParams['font.family'] ='AppleGothic' # For MacOS , Windows : 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.figure(figsize=(7,4))
plt.hist(lengths, bins=20, color="lightgreen", edgecolor="black")
plt.xlabel("서열 길이")
plt.ylabel("빈도수")
plt.title("서열 길이 분포")
plt.tight_layout()
plt.show()
```

```python
# FASTA file load and parsing
fasta_file = "output.fasta"
records = list(SeqIO.parse(fasta_file, "fasta"))

analysis_results = []

for record in records:
    seq_id = record.id
    sequence = record.seq

    # 서열 길이 및 GC 비율 계산
    seq_length = len (sequence)
    gc_content = gc_fraction(record.seq) * 100

    analysis_results.append({
        "ID": seq_id,
        "Length": seq_length,
        "GC_Content(%)": gc_content
    })
```

```python
print(analysis_results[:5])  # 처음 5개 결과 출력
print(f"{'유전자 ID':<15} {'서열 길이 (bp)':<15} {'GC 비율 (%)':<15}")
print("-"* 400)
for result in analysis_results:
    print(f"{result['ID']:<20}{result['Length']:<20}{result['GC_Content(%)']:.2f}")
```

```python
gene_ids = [result['ID'] for result in analysis_results]
gc_contents = [result['GC_Content(%)'] for result in analysis_results]

lengths = [result['Length'] for result in analysis_results]

# GC 비율 시각화
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(gene_ids, gc_contents, color='skyblue', edgecolor='black')
plt.xlabel("Gene ID")
plt.ylabel("GC Content (%)")
plt.title("GC Content of Genes")
plt.xticks(rotation=45)


# 서열 길이 시각화
plt.subplot(1, 2, 2)
plt.bar(gene_ids, lengths, color='lightgreen', edgecolor='black')
plt.xlabel("Gene ID")
plt.ylabel("Sequence Length")
plt.title("Sequence Length of Genes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
